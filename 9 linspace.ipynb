#IF YOU WANT EQUALLY SPACED ARRAY ELEMENTS USE .LINSPACE

```python
import numpy as np

arr_random=np.linspace(0,10,num=5) #start from 0 and include all number upto 10 by dividing it into 5 parts.
print(arr_random)
```

#start array from 0 upto 30. the 3rd paramter is the number of element you want in the array. if you set third paramter 3, you will get only 3 elements that include starting term, ending term and the middle term.

```python
arr_random=np.linspace(0,30,3)
print(arr_random)
```
