#DIAGONAL.
IF YOU WANT 1 in diagonal use .eye 

```python
import numpy as np

arr_diagonal=np.eye(3)
print(arr_diagonal)

print(f"the dimesnion is {arr_diagonal.ndim}")

#here you will get 3x3 matrix with diagonal elements 1. it automatically creates a 2D Array.
```
