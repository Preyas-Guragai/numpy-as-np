```python
# .empty function


import numpy as np

arr_empty=np.empty(4)
print(arr_empty)         #we get random value or the value that is already in that memory.

print(type(arr_empty))
print(arr_empty.ndim)
```

```python
#explicitely passing rows and columns

array_empty=np.empty((3,4))   # 3 rows and 4 columns with random values

print(array_empty)
print(f"the dimension of the empty array is {array_empty.ndim}")
```
