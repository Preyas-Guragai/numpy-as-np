#Arithmetic operations in 2D Array

#2d array multiplication using np function and without using functions

```python
import numpy as np

var=np.array([[1,2,3],[4,5,6]])
result=var*2
print(result)
```

```python
import numpy as np

var=np.array([[1,2,3],[4,5,6]])
result=np.multiply(var,2)
print(result)
```

```python
var1=np.array([[1,2,3],[4,5,6]])
var2=np.array([[1,2,3],[4,5,6]])
result=np.multiply(var1,var2)
print(result)
```

#adding 2d np array with and without np functions

```python
var=np.array([[1,2,3],[4,5,6]])
result=var+5 #without using np function
print(result)
```

```python
var=np.array([[1,2,3],[4,5,6]])
result=np.add(var,5) #with using np function
print(result)
```
