#rand(); used to generate number between 0 and 1 and all +ve numbers. [0,1)

```python
import numpy as np

variable=np.random.rand(4)      #the parameter says how many random values in the array.
print(variable)
print(f"dimension is {variable.ndim}")
print(type(variable))


#note: everytime you run this cell, you will see the differnt values in the n-d array
```

#explicitely pass array order i*j

```python
var=np.random.rand(3,4)   #3 rows and 4 column.
print(var)
```

#randn(); this will generate value close to 0 that may include +ve and -ve as well


```python
var=np.random.randn(5)
print(var)
```

```python
var=np.random.randn(2,3)
print(var)
```

#ranf(). similar to rand(). [0,1) range.

```python
var=np.random.ranf(4)
print(var)
print(var.ndim)
print(type(var))
```

#RANDINT(). it generates random value in the given range by user.


```python
#np.random.randint(min_value,max_value, total_values

var=np.random.randint(5,20,5)

print(var)
```
