                                       #ARITHMETIC OPERATIONS

#ADD

```python
import numpy as np

var=np.array([1,2,3,4,5])
var+=10
print(var)
```

```python
var1=np.array([1,2,3,4,5])
var2=np.array([1,2,3,4,5])
varAdd=var1+var2
print(varAdd)
```

#subtract

```python
var=np.array([1,2,3,4,5])
varSub=var-10
print(varSub)
```

```python
var1=np.array([1,2,3,4,5])
var2=np.array([1,2,3,4,5])
varSub=var1-var2
print(varSub)
```

#divide

```python
var=np.array([1,2,3,4,5])
div=var/10
print(div)
```

```python
var1=np.array([10,20,30,40,50])
var2=np.array([5,4,3,2,1])
varDiv=var1/var2
print(varDiv)
```

#multiply

```python
var=np.array([1,2,3,4,5])
mul=var*10
print(mul)
```

```python
var1=np.array([10,20,30,40,50])
var2=np.array([5,4,3,2,1])
varMul=var1*var2
print(varMul)
```

#modulus

```python
var=np.array([11,20,33,40,55])
modulus=var%10
print(modulus)
```

```python
var1=np.array([10,20,30,40,50])
var2=np.array([3,4,2,15,1])
varMod=var1%var2
print(varMod)
```
