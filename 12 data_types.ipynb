#DATA TYPES

```python
import numpy as np

var=np.array([1,2,3,4])
print("Data types:", var.dtype)
```

```python
var=np.array([1.34, 3.45, 8.90])
print("data type: ", var.dtype)
```

```python
var=np.array(["a","b","c"])
print("data type:", var.dtype)
```

```python
var=np.array([1,"b","c",2])
print("data type:", var.dtype)
```

```python
var=np.array([1,2,3.5,6.23])
print("data type:", var.dtype)

#note: here the data is int and float. np array cannot keep different data in one array,
#....so it upcasts everything to the most general that can hold all values. here float is more general.
```

#FIRST CELL EXAMPLE IS INT64. IT CAN BE CHANGED TO INT8. LET'S SEE HOW??

```python
import numpy as np

var=np.array([1,2,3,4],dtype=np.int8)
print("Data types:", var.dtype)
```

#INT INTO FLOAT USING SYMBOL OF FLOAR i.e "f"

```python
import numpy as np

var=np.array([1,2,3,4],dtype="f")
print("Data types:", var.dtype)
```

#changing int into float and again into int

```python
import numpy as np

var=np.array([1,2,3,4])
var1=np.float32(var)
var2=np.int_(var1)


print("Data types:", var.dtype)
print("Data types:", var1.dtype)
print("Data types:", var2.dtype)


print(var)
print(var1)
print(var2)
```

#automatically convert data type using .astype()

```python
var=np.array([1,2,3,4])
var1=var.astype(float)
print(var1.dtype)
```
