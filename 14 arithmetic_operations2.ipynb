                                       #ARITHMETIC OPERATIONS with np function

#ADD --- np.add(parameter1, parameter2)

```python
import numpy as np

var=np.array([1,2,3,4,5])
add=np.add(var,3)
print(add)
```

```python
var1=np.array([1,2,3,4,5])
var2=np.array([1,2,3,4,5])
varAdd=np.add(var1,var2)
print(varAdd)
```

#SUBTRACT---- np.subtract(var1,var2)

```python
var=np.array([1,2,3,4,5])
varSub=np.subtract(var,3)
print(varSub)
```

```python
var1=np.array([1,2,3,4,5])
var2=np.array([2,3,4,5,6])
varSub=np.subtract(var1,var2)
print(varSub)
```

#DIVIDE

```python
var=np.array([3,6,9,12])
div=np.divide(var,3)
print(div)
```

```python
var1=np.array([10,20,30,40,50])
var2=np.array([5,4,3,2,4])
varDiv=np.divide(var1,var2)
print(varDiv)
```

#multiply--- np.multiply(x,x)


```python
var1=np.array([10,20,30,40,50])
var2=np.array([5,4,3,2,4])
varMul=np.multiply(var1,var2)
print(varMul)
```

#MODULUS --np.mod(x,x)

```python
var1=np.array([10,20,30,40,50])
var2=np.array([5,3,9,2,4])
varMod=np.mod(var1,var2)
print(varMod)
```

```python
import numpy as np

var=np.array([1,2,3,4,5])
mod=np.mod(var,3)
print(mod)
```

#power----np.power(var,power)

```python
import numpy as np

var=np.array([1,2,3,4,5])
pow=np.pow(var,3)
print(pow)
```

```python
var1=np.array([1,2,3,4])
var2=np.array([1,2,3,2])
varPow=np.power(var1,var2)
print(varPow)
```

#RECIPROCAL ----np.reciprocal(var1)

```python
import numpy as np

var=np.array([1,20,0.5,1/2,5],dtype="f")
reciprocal=np.reciprocal(var)
print(reciprocal)
```

```python
var=np.array([1,2,3,4,5])
reciprocal=np.reciprocal(var)
print(reciprocal)

#this array is int so result is only given in integer form
```
